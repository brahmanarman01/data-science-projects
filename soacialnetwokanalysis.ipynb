```python


def load_data(filename):
  with open(filename, 'r') as f:
    return json.load(f)

def find_people_you_may_know(user_id,data):
  user_freinds = {}
  for user in data['users']:
    user_freinds[user['id']] = set(user['friends'])
  if user_id not in user_freinds:
    return []

  direct_friends = user_freinds[user_id]
  suggestions ={}
  for friend in direct_friends:
    for muutual in user_freinds[friend]:
      if muutual != user_id and muutual not in direct_friends:
        if muutual not in suggestions:
            suggestions[muutual] = suggestions.get(muutual,0) + 1
  sorted_suggestions = sorted(suggestions.items(), key=lambda x: x[1], reverse=True)
  return [user_id for user_id, mutual_count in sorted_suggestions[:5]]
data = load_data("massive_data.json")

## Find people you may know
user_id = 5
recc = find_people_you_may_know(user_id,data)
print(recc)
```
